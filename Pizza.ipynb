### **PIZZA RECOMMENDATION AND IT'S ANALYSIS**

This is a list of pizzas from two of the cities in the country of us namely Los Angeles and Lawrence Township.It helps in predicting least,and expensive restaurants and cities for pizza.
and it is also helpful for knowing the most selled pizzas and least selled pizzas.

```python
import pandas as pd
import matplotlib. pyplot as plt
d=pd.read_csv("/content/drive/MyDrive/Pizza.csv")
d.head()
```

```python
x=d["city"]
y=d["menus.amountMax"]
plt.bar(x,y)
plt.xlabel("City")
plt.ylabel("Maximum price of pizza")
plt.show()
```

It can be observed that the cost of pizza is bit more in LosAngeles when compared to Lawrence Township


```python
x=d["city"]
y=d["menus.amountMin"]
plt.xlabel("City")
plt.ylabel("Minimum price of pizza")
plt.bar(x,y)
plt.show()
```


It can be observed that the cost of pizza is at cheaper cost in Lawrence Township when compared to LosAngeles 



```python
x=d["menus.dateSeen"]
y=d["priceRangeMin"]
plt.xlabel("Month")
plt.ylabel("Minimum price ")
plt.bar(x,y)
plt.show()
```

It can be observed that these sales are done in respective months.

```python
x=d["menus.dateSeen"]
y=d["priceRangeMax"]
plt.xlabel("Month")
plt.ylabel("Maximum price of pizza")
plt.bar(x,y)
plt.show()
```

It can be observed that no sales of pizza done in the month of december-2021 and the highest sales are done in the month of march-2022

```python
x=d["menus.amountMax"]
y=d["menus.name"]
plt.xlabel("Maximum price of pizza",fontsize=20)
plt.ylabel("Varieties of pizza",fontsize=20)
plt.scatter(x,y)
plt.show()
```

It can be observed that the largest only single pizza variety named Shrimp Alfredo Pizza is having the maximum cost i.e around 26.99$

```python
x=d["menus.amountMin"]
y=d["menus.name"]
plt.xlabel("Minimum price of pizza",fontsize=20)
plt.ylabel("Varieties of pizza",fontsize=20)
plt.scatter(x,y)
plt.show()
```

It can be observed that  the cheapest pizza among all the varieties is cheese slice pizza with the minimal cost of 2$ available at lawrence township pizza centre.
